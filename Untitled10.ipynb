```python
#EXP 1

import numpy as np

class Perceptron:
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        self.w = np.zeros(X.shape[1])
        self.b = 0
        y = np.where(y <= 0, -1, 1)

        for _ in range(self.epochs):
            for x, target in zip(X, y):
                pred = np.sign(np.dot(x, self.w) + self.b)
                if pred != target:
                    self.w += self.lr * target * x
                    self.b += self.lr * target

    def predict(self, X):
        return np.sign(np.dot(X, self.w) + self.b)


# Example
X = np.array([[1,2],[2,3],[3,4],[1,0],[0,1],[3,1]])
y = np.array([1,1,1,-1,-1,-1])

p = Perceptron()
p.fit(X, y)

print("Predicted:", p.predict(X))
print("Actual:   ", y)
```

```python
#EXP 2

import numpy as np

class FFNN:
    def __init__(self):
        self.lr = 0.1
        self.w1 = np.random.randn(1, 10) * 0.1
        self.b1 = np.zeros((1, 10))
        self.w2 = np.random.randn(10, 1) * 0.1
        self.b2 = np.zeros((1, 1))

    def sigmoid(self, x):
        return 1 / (1  + np.exp(-x))

    def d_sigmoid(self, x):
        return x * (1 - x)

    def forward(self, X):
        self.h = self.sigmoid(np.dot(X, self.w1) + self.b1)
        self.o = np.dot(self.h, self.w2) + self.b2
        return self.o

    def train(self, X, y, epochs):
        for i in range(epochs):
            out = self.forward(X)

            error = y - out
            d_out = -2 * error

            h_error = np.dot(d_out, self.w2.T)
            d_h = h_error * self.d_sigmoid(self.h)

            self.w2 -= self.lr * np.dot(self.h.T, d_out)
            self.b2 -= self.lr * np.sum(d_out, axis=0)

            self.w1 -= self.lr * np.dot(X.T, d_h)
            self.b1 -= self.lr * np.sum(d_h, axis=0)

            if i % 100 == 0:
                print(f"Epoch {i}, Loss: {np.mean(error**2)}")

# Data
X = np.array([[0], [1], [2], [3], [4]], dtype=float)
y = np.array([[0], [2], [4], [6], [8]], dtype=float)

# Scaling
X /= np.max(X)
y /= np.max(y)

# Train
nn = FFNN()
nn.train(X, y, 1000)

# Prediction
pred = nn.forward(X)

print("Predictions:", pred)
print("Actual values:", y)
```

```python
#EXP 3

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report
import numpy as np

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

X_train = X_train.reshape(60000, 784) / 255.0
X_test = X_test.reshape(10000, 784) / 255.0

model = Sequential([
    tf.keras.Input(shape=(784,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=10,
    validation_split=0.2
)

train_pred = np.argmax(model.predict(X_train), axis=1)
test_pred = np.argmax(model.predict(X_test), axis=1)

print("\nModel Summary")
model.summary()

print("\nTraining Data Evaluation")
print(classification_report(y_train, train_pred))

print("\nTest Data Evaluation")
print(classification_report(y_test, test_pred))
```

```python
#EXP 4

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize & reshape
X_train = X_train.reshape(-1, 784) / 255.0
X_test = X_test.reshape(-1, 784) / 255.0

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Model
model = Sequential([
    Input(shape=(784,)),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='softmax')
])

# Compile
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(
    X_train, y_train,
    epochs=10,
    validation_data=(X_test, y_test),
    verbose=1
)

# Plot
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Evaluate
loss, acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", acc)
```

```python
#EXP 6

import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.optimizers import Adam
import numpy as np

# Dataset path
train_dir = "D:/img_ip"

# Data preprocessing
data = ImageDataGenerator(rescale=1./255)

train = data.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=2,
    class_mode='categorical'
)

# Load VGG16
base_model = VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

# Freeze layers
for layer in base_model.layers:
    layer.trainable = False

# Build model
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

# Compile
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train
model.fit(train, epochs=10)

# Predict image
img_path =  "D:/img_test/download.jpg"

img = load_img(img_path, target_size=(224,224))
img = img_to_array(img) / 255.0
img = np.expand_dims(img, axis=0)

pred = np.argmax(model.predict(img))

labels = ['apples', 'tomatoes']

print("Predicted Class:", labels[pred])
```

```python
#EXP 7

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, LSTM, RepeatVector
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import mnist

# Load data
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Model
inp = Input(shape=(28,28))
x = LSTM(32)(inp)
x = RepeatVector(28)(x)
out = LSTM(28, return_sequences=True)(x)

model = Model(inp, out)

model.compile(optimizer='adam', loss='mse')

# Train
model.fit(
    x_train, x_train,
    epochs=5,
    batch_size=128,
    validation_data=(x_test, x_test)
)

# Predict
decoded = model.predict(x_test[:10])

# Display
plt.figure(figsize=(15,4))

for i in range(10):
    # Original
    ax = plt.subplot(2,10,i+1)
    plt.imshow(x_test[i], cmap='gray')
    ax.axis('off')

    # Reconstructed
    ax = plt.subplot(2,10,i+11)
    plt.imshow(decoded[i], cmap='gray')
    ax.axis('off')

plt.show()
```

```python
# EXP 8

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Reshape
from tensorflow.keras.datasets import mnist

# Load MNIST
(X_train, _), (_, _) = mnist.load_data()

# Normalize
X_train = X_train / 255.0

# Generator model
generator = Sequential([
    Dense(128, activation='relu', input_dim=100),
    Dense(28 * 28, activation='sigmoid'),
    Reshape((28,28))
])

generator.compile(optimizer='adam', loss='mse')

# Random noise
noise = np.random.normal(0, 1, (X_train.shape[0], 100))

# Train
history = generator.fit(
    noise,
    X_train,
    epochs=5,
    batch_size=128
)

# Plot history
plt.plot(history.history['loss'])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.show()

# Generate 5 images
test_noise = np.random.normal(0,1,(5,100))

imgs = generator.predict(test_noise)

# Show images
plt.figure(figsize=(10,2))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(imgs[i], cmap='gray')
    plt.axis('off')

plt.show()
```

```python
#EXP 9 

import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load data
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)

# Pad sequences
X_train = pad_sequences(X_train, maxlen=500)
X_test = pad_sequences(X_test, maxlen=500)

# Model
model = Sequential([
    Embedding(10000, 128),
    Conv1D(32, 5, activation='relu'),
    MaxPooling1D(2),
    Conv1D(64, 5, activation='relu'),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=128,
    validation_data=(X_test, y_test)
)

# Evaluate
loss, acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", acc)

# Plot
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#EXP 10

import tensorflow as tf
import numpy as np

# Load text
path = tf.keras.utils.get_file(
    "shakespeare.txt",
    "https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt"
)

text = open(path, 'rb').read().decode('utf-8')

# Vocabulary
vocab = sorted(set(text))
char2idx = {u:i for i,u in enumerate(vocab)}
idx2char = np.array(vocab)

text_as_int = np.array([char2idx[c] for c in text])

# Dataset
seq_len = 100

data = tf.data.Dataset.from_tensor_slices(text_as_int)

seq = data.batch(seq_len+1, drop_remainder=True)

def split(chunk):
    return chunk[:-1], chunk[1:]

dataset = seq.map(split).shuffle(10000).batch(64, drop_remainder=True)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(vocab), 256),
    tf.keras.layers.LSTM(512, return_sequences=True),
    tf.keras.layers.Dense(len(vocab))
])

# Compile
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer='adam', loss=loss)

# Train 
model.fit(dataset, epochs=1)

# Text generation
def generate(start="ROMEO: ", num=300):

    inp = tf.expand_dims([char2idx[c] for c in start], 0)

    text_gen = []

    for _ in range(num):

        pred = model(inp)

        pred = tf.squeeze(pred, 0)

        pred_id = tf.random.categorical(pred[-1:], 1)[0,0].numpy()

        inp = tf.expand_dims([pred_id], 0)

        text_gen.append(idx2char[pred_id])

    return start + ''.join(text_gen)

# Output
print(generate())
```
